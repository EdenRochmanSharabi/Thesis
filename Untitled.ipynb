```python
import pandas as pd
import matplotlib.pyplot as plt
df_results = pd.read_csv('total_utilities_by_capacity.csv')
# Drop the 'Capacity' and 'Optimization' columns and calculate the mean utilities for each solver across all capacities
solver_averages = df_results.drop(columns=['Capacity', 'Optimization']).mean().sort_values(ascending=False)

# Define a modern, soft blue color
modern_blue = '#4A90E2'

# Plot with enhanced resolution and Apple-style design
plt.figure(figsize=(12, 8), dpi=150)
plt.bar(solver_averages.index, solver_averages.values, color=modern_blue)
plt.grid(axis='y', linestyle='--', color='#D3D3D3', alpha=0.5)
plt.xlabel('Solvers', fontsize=14, fontweight='medium', color='#333333')
plt.ylabel('Average Utility Across All Capacities', fontsize=14, fontweight='medium', color='#333333')
plt.title('Average Utility for Each Solver Across All Capacities', fontsize=16, fontweight='light', color='#333333')
plt.xticks(rotation=45, fontsize=12, color='#333333')
plt.yticks(fontsize=12, color='#333333')
plt.tight_layout()

# Set plot aesthetics
plt.gca().set_facecolor('white')
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['left'].set_visible(False)
plt.gca().spines['bottom'].set_color('#DDDDDD')

plt.show()
```
